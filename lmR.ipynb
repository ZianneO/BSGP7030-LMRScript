# Linear Modeling in R

### Reading in the csv file

```r
df = read.csv("regrex1.csv")
```

```r
is.data.frame(df)
```

```r
df
```

```r
head(df)
```

```r
str(df)
```

# Scatter Plot of data

```r
plot(df$x, df$y)
```

# Linear Model

```r
lm(y~x, data = df)
```

```r
LMR <- lm(y~x, data = df)
```

# Summary of Linear Model

```r
summary(LMR)
```

```r
summary_stats = summary(LMR)
```

# Obtaining Coefficients
### R^2

```r
summary_stats$r.squared
```

### Intercept and Slope

```r
coef(LMR)
```

```r
coef(LMR)[1]
```

```r
coef(LMR)[2]
```

```r
y_predict = predict(LMR, df)
```

# Linear Regression of Scatter Plot

```r
plot(df$x, y_predict)
```

# Linear Regression Plot Line Addition

```r
plot(df$x, df$y)
lines(df$x, y_predict)
```

# Title Addition

```r
plot(df$x, df$y,
     main = "Scatter Plot with Linear Regression Line")
lines(df$x, y_predict)
```

# Axes Addition

```r
plot(df$x, df$y,
     xlab = "", ylab = "",
     main = "Scatter Plot with Linear Regression Line")
lines(df$x, y_predict)
xlabel <- "x-axis"
ylabel <- "y-axis"
title(xlab = xlabel, col.lab = "pink")
title(ylab = ylabel, col.lab = "purple")

abline(LMR, col = "blue")
```

# Slope and R^2 annotations

```r
plot(df$x, df$y,
     xlab = "", ylab = "",
     main = "Scatter Plot with Linear Regression Line")
lines(df$x, y_predict)
xlabel <- "x-axis"
ylabel <- "y-axis"
title(xlab = xlabel, col.lab = "pink")
title(ylab = ylabel, col.lab = "purple")

abline(LMR, col = "blue")
slope_annotation <- sprintf("Slope: %.2f", LMR$coefficients[2])
text(x = max(df$x) * 0.9, y = max(df$y) * 0.9, label = slope_annotation, col = "red")
R2_annotation <- sprintf("R^2: %.2f", summary(LMR)$r.squared)
text(x = max(df$x) * 0.9, y = max(df$y) * 0.85, label = R2_annotation, col = "red", adj = 0)
```
